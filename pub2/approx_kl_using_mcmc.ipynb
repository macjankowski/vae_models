```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
import scipy.stats
from scipy.stats import multivariate_normal
```

```python
def plot_normal(mean, cov):
    x, y = np.random.multivariate_normal(mean, cov, 5000).T
    plt.plot(x, y, 'x')
    plt.axis('equal')
    plt.tight_layout()
    #plt.axis('off')
    #plt.axes().set_aspect('equal', 'datalim')
    plt.show()
```

```python
def get_sample(params):
    s = np.random.multivariate_normal(params.mean, params.cov, 1)[0].T
    return s
```

```python
def generate_samples_from_gmm(norm_params, weights, N):
    mixture_idx = np.random.choice(len(weights), size=N, replace=True, p=weights)
    samples = [get_sample(norm_params[i]) for i in mixture_idx]
    return np.array(samples)
```

```python
class NormParams():
    
    def __init__(self, mean, cov):
        self.mean = mean
        self.cov = cov
```

```python
def pdf_mix(x, means, stdevs, weights):
    a = scipy.stats.norm.pdf(x, loc=means[0], scale=stdevs[0])
    b = scipy.stats.norm.pdf(x, loc=means[1], scale=stdevs[1])
    return weights[0]*a + weights[1]*b
```

```python
def estimate_kl(from_mean, from_std, mix_means, mix_stds, mix_weigths, N):
    s = scipy.stats.norm.rvs(loc=from_mean, scale=from_std, size=N)
    return np.mean(np.log(scipy.stats.norm.pdf(s, loc=from_mean, scale=from_std)/pdf_mix(s, mix_means, mix_stds, mix_weigths)))
```

```python
def pdf_mix_mult(x, params, weights):
    margins = [norm_dens.pdf(x) for norm_dens in params]
    dot_prod = np.sum(np.array(margins).T*weights, axis=1)
    return dot_prod
```
